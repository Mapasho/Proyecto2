```python
import pandas as pd 
import numpy as np
```

```python
df= pd.read_csv('AccidentesAviones.csv', index_col=0)
```

```python
df.head(20)
```

# se elimina flight, route, registration por una cuestion de criterio el cual se observo durante 3 dias 

```python
df2= df.drop(columns=['flight_no', 'route', 'registration'])
```

```python
df2.head()
```

```python
df2.replace("?",np.nan, inplace=True) # se reemplazan los signos por nulos
```

```python
df2[['mes','dia','año']] = df2['fecha'].str.split(' ', expand=True) # se crea las columnas dia, mes, anio para desanidar
```

```python
df2['dia'] = df2.dia.str.replace(',', '')
```

```python
df2['mes'] = pd.to_datetime(df2['mes'], format= '%B').dt.month
```

```python
df2['fecha'] = df2['año'].astype(str)+'-'+df2['mes'].astype(str)+'-'+df2['dia'].astype(str)
```

```python
df2['fecha'].value_counts
```

```python
# df2.drop(columns=['mes','dia'], inplace=True)
```

```python
df2.head()
```

```python
df2[df2['HORA declarada'].str.len()>4]
```

```python
df2['HORA declarada'] = df2['HORA declarada'].str.replace(':', '')
```

```python
df2['HORA declarada'] = df2['HORA declarada'].str.replace('c', '')
```

```python
df2['HORA declarada'] = df2['HORA declarada'].str.replace('Z', '')
```

```python
df2['HORA declarada'] = df2['HORA declarada'].str.replace(';', '')
```

```python
df2['HORA declarada']= df2['HORA declarada'].str.strip()
```

```python
df2['hora']=pd.to_datetime(df2['HORA declarada'], format='%H%M').dt.time
```

```python
df2.head()
```

```python
df2.drop(columns='HORA declarada', inplace=True)
```

```python
df2
```

```python
df2.rename(columns={'OperadOR': 'operador', 'PASAJEROS A BORDO': 'pasajeros_a_bordo', 'crew_aboard': 'personal_a_bordo', 'all_aboard': 'total_a_bordo',
           'passenger_fatalities': 'pasajeros_fallecidos', 'crew_fatalities': 'personal_fallecido', 'cantidad de fallecidos': 'total_fallecidos',
           'ac_type': 'tipo_de_avion','Ruta': 'ruta'}, inplace=True)
```

```python
df2 = df2.reindex(columns=['fecha', 'hora', 'operador','tipo_de_avion','ruta','pasajeros_a_bordo','personal_a_bordo','total_a_bordo','pasajeros_fallecidos','personal_fallecido','ground','total_fallecidos']
                 + df2.columns.difference(['fecha', 'hora','tipo_de_avion','ruta' ,'operador', 'pasajeros_a_bordo','personal_a_bordo','total_a_bordo','pasajeros_fallecidos','personal_fallecido','ground','total_fallecidos']).tolist())
```
